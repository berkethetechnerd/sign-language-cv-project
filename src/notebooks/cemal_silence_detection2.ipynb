```python
test_dir = './output_video_segmentation/'
test_imgs = ['./output_video_segmentation/{}'.format(i) for i in os.listdir(test_dir)]  # get test images
```

```python
nrows = 150
ncolumns = 150
channels = 3
```

```python
# A function to read and process the images in an acceptable format for our model
# Returns two arrays: X -> resized images, y -> labels
def read_and_process_image(list_of_images):
    x = []
    y = []
    
    for image in list_of_images:
        x.append(cv2.resize(cv2.imread(image, cv2.IMREAD_COLOR), (nrows, ncolumns), interpolation=cv2.INTER_CUBIC)) # Read the image
        if 'down' in image:
            y.append(1)
        elif 'up' in image:
            y.append(0)
    
    return x,y
```

```python
from keras import layers
from keras import models
from keras import optimizers
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing.image import img_to_array, load_img
```

```python

from keras.models import load_model
model = load_model('./model_keras.h5')
```

```python
import shutil
from shutil import copy

# Define the output directory
test_result_up = './test_result_up/'
test_result_down = './test_result_down/'

# Clear if the output directory exists
if os.path.exists(test_result_up):
    shutil.rmtree(test_result_up, ignore_errors = True)
    
if os.path.exists(test_result_down):
    shutil.rmtree(test_result_down, ignore_errors = True)

# Create the output directory
os.mkdir(test_result_up)
os.mkdir(test_result_down)
up = 1384;
down = 1480;
```

```python
from PIL import Image
import numpy as np

for i in range(len(test_imgs)):
    x_test, y_test = read_and_process_image(test_imgs[i:i+1])
    x = np.array(x_test)
    test_datagen = ImageDataGenerator(rescale=1./255)

    for batch in test_datagen.flow(x, batch_size=1):
        pred = model.predict(batch)

        if pred > 0.5:
            copy(test_imgs[i], './test_result_down/down_{}.jpg'.format(down) )
            down+=1
        else:
            copy(test_imgs[i], './test_result_up/up_{}.jpg'.format(up) )
            up+=1

        i+=1
        
        if(i%33==0 or i==len(test_imgs)):
            print('{perc}& of process is completed.'.format(perc=i/len(test_imgs)*100))
        
        break
```
